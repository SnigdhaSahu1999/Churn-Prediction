```python
cdf['tenure']=cdf['tenure'].replace({0:0.1})
ten=np.log(cdf['tenure'])
cdf.drop(["tenure"],axis=1,inplace=True)
cdf=pd.concat((cdf,ten),axis=1)
```

```python
from sklearn.model_selection import train_test_split
x=cdf.drop('Churn',axis=1)
y=cdf['Churn']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel= LogisticRegression()
logmodel.fit(x_train,y_train)
pre= logmodel.predict(x_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
print()
print("Accuracy of the logistic model:  " , accuracy_score(y_test, pre))
print("F1_score of the model: ", f1_score(y_test,pre))
print()
```

```python
from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,pre)
print(res)
print()
tn,tp,fn,fp=confusion_matrix(y_test,pre).ravel()
print(tn,tp,fn,fp)
```
